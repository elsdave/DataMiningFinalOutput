# Bivariate Analysis üöÄ‚ú®

```python
features = [
    'GENERAL APPEARANCE', 'MANNER OF SPEAKING', 'PHYSICAL CONDITION',
    'MENTAL ALERTNESS', 'SELF-CONFIDENCE', 'ABILITY TO PRESENT IDEAS',
    'COMMUNICATION SKILLS', 'Student Performance Rating'
]

df['EMPLOYABILITY_SCORE'] = df[features].mean(axis=1)
df['EMPLOYABILITY_SCORE']
```

```python
print(df.head())
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score
```

```python
X = df.drop(columns = ['Name of Student', 'CLASS', 'Class_Encoded'])
y = df['CLASS']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.1, random_state=42)
```

```python
catboost_model = CatBoostClassifier(iterations=500, learning_rate=0.01, depth=6, verbose=False, random_state=42, bagging_temperature=2, l2_leaf_reg=7 )
catboost_model.fit(X_train, y_train)
```

```python
y_pred = catboost_model.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
print(f'Accuracy: {accuracy:.4f}')
print(f'Percentage Accuracy: {accuracy*100}%')
```

```python
classification_report_catboost = classification_report(y_test, y_pred)
print(classification_report_catboost)
```
